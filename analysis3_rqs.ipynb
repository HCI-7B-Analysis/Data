```python
import polars as pl

RQs = ["RQ1", "RQ2", "RQ3"]

# Get combined 
studies = (
    pl.read_csv("./combined.csv")
        .with_columns(RQs=pl.col("Tags").str.split("; "))
        .filter(pl.col("category").ne(pl.lit("Review")))
)

studies_per_rq = {
    rq: studies.filter(studies["RQs"].list.contains(pl.lit(rq))) for rq in RQs
}
```

```python
studies.filter(pl.col("paper_id").is_in(pl.lit([10]))).select(pl.col("paper_id", "Title", "Authors", "Published Year"))
```
